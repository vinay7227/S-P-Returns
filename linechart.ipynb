```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_spx_ftse_00_10 = pd.read_csv("line_chart_data.csv")
```

```python
df_spx_ftse_00_10 
```

```python
#I converted the date string into a time and date format
df_spx_ftse_00_10["new_date"] = pd.to_datetime(df_spx_ftse_00_10["Date"])
df_spx_ftse_00_10["new_date"]
```

```python
labels = ["S&P 500", "FTSE100" ]
plt.figure(figsize=(20, 8))
plt.plot(df_spx_ftse_00_10["new_date"], df_spx_ftse_00_10["GSPC500"])
plt.plot(df_spx_ftse_00_10["new_date"], df_spx_ftse_00_10["FTSE100"])
plt.title("S&P vs FTSE Returns(2000-2010)", fontsize= 14, fontweight="bold")
plt.xlabel("Date", fontsize=16)
plt.ylabel("Returns", fontsize=16)
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
#To enhance granularity, I have selected the specific time frame of the second quarter of 2008.
df_spx_ftse_08 = df_spx_ftse_00_10[(df_spx_ftse_00_10.new_date>='2008-07-01') &(df_spx_ftse_00_10.new_date<='2008-12-31')]
df_spx_ftse_08
```

```python
labels = ["S&P 500", "FTSE100" ]
plt.figure(figsize=(30, 10))
plt.plot(df_spx_ftse_08["new_date"], df_spx_ftse_08["GSPC500"], color = "midnightblue")
plt.plot(df_spx_ftse_08["new_date"], df_spx_ftse_08["FTSE100"], color = "crimson")
plt.title("S&P vs FTSE Returns(2nd quarter 2008)", fontsize= 14, fontweight="bold")
plt.xlabel("Date", fontsize=16)
plt.ylabel("Returns", fontsize=16)
plt.legend(labels = labels, fontsize = "large")
plt.show()
```
